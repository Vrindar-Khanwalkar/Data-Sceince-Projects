Fill Null Values, If a particular feature has lot of null values, See the impact and make a decision to either drop it.

```python
data['Age'].fillna(data['Age'].median(),inplace=True)

data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)

data.drop(columns=['Cabin'],inplace= True)

```

```python
data
```

Feature Creation - Create meaningful feature for your data to make sense and make the analysis easy.


```python
all_titles = [' Capt', ' Col', ' Don', ' Dr', ' Jonkheer', ' Lady', ' Major',' Master', ' Miss', ' Mlle', ' Mme', ' Mr', ' Mrs', ' Ms', ' Rev', ' Sir', ' the Countess']

data['Title'] = data['Name'].str.extract(' ([A-Za-z]+)\.', expand=False) # Regex for Name : Braund, Mr. Owen Harris -- Mr is selected.

# Simplify rare titles into one category
rare_titles = ['Lady', 'Countess','Capt', 'Col','Don','Dr','Major','Rev','Sir','Jonkheer','Dona']
data['Title'] = data['Title'].replace(rare_titles, 'Rare')
data['Title'] = data['Title'].replace({'Mlle':'Miss', 'Ms':'Miss', 'Mme':'Mrs'})
```

```python
data['Family_Size'] = data["SibSp"] + data['Parch'] + 1
```

```python
data['IsAlone'] = 1  # Initialize to 1 (alone)
data['IsAlone'].loc[data['Family_Size'] > 1] = 0
```

```python
data.loc[data['Age'] <= 16, 'AgeGroup'] = 0
data.loc[(data['Age'] > 16) & (data['Age'] <= 32), 'AgeGroup'] = 1
data.loc[(data['Age'] > 32) & (data['Age'] <= 48), 'AgeGroup'] = 2
data.loc[(data['Age'] > 48) & (data['Age'] <= 64), 'AgeGroup'] = 3
data.loc[data['Age'] > 64, 'AgeGroup'] = 4
```

```python
data['AgeBand'] = pd.cut(data['Age'],5)
```

```python
data
```

```python
data['Sex'] = data['Sex'].map({'male':0,'female':1})
```

```python
# One-hot encode 'Embarked'
embarked_dummies = pd.get_dummies(data['Embarked'], prefix='Embarked')
data = pd.concat([data, embarked_dummies], axis=1)

# One-hot encode 'Title'
title_dummies = pd.get_dummies(data['Title'], prefix='Title')
data = pd.concat([data, title_dummies], axis=1)

```

```python
data
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
data['Fare_scaled'] = scaler.fit_transform(data[['Fare']])
data['Age_scaled'] = scaler.fit_transform(data[['Age']])
```

```python
ageband_dummies = pd.get_dummies(data['AgeBand'], prefix='AgeBand')
data = pd.concat([data, ageband_dummies], axis=1)
```

```python
print(data.dtypes)
```

```python
from sklearn.ensemble import RandomForestClassifier

X = data.drop(['Survived', 'Name', 'Ticket', 'Embarked', 'Title','AgeBand'], axis=1)
y = data['Survived']

model = RandomForestClassifier()
model.fit(X, y)

# Feature importance
importances = model.feature_importances_
feature_names = X.columns
importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
print(importance_df.sort_values(by='Importance', ascending=False))
```
